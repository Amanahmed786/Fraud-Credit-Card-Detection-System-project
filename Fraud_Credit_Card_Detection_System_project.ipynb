```python
#importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn .model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to pandas dataframe
credit_card_data = pd.read_csv('creditcard.csv')
```

This dataset is highly unbalanced


0->Normal transaction
1->fraud transaction

```python
# seperating the data for analyses
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

#statistical measure of the data


```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

Under sampling

Build a sample dataset containing similar distribution of normal transaction and fraud transaction

Number of Fraudulent Transactions-->492

```python
legit_sample = legit.sample(n=492)
```

Concating two Dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features & Targets

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
print(x)
```

```python
print(y)
```

Split the data into Training data & Testing



Approch 1

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
#traing the logistic regression Model with Training Data
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on train data:',training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on teat data:',test_data_accuracy)
```
